#Lendo vários arquivos

#Importando Bibliotecas

```python
import pandas as pd
from glob import glob
```

#Lendo Arquivos

```python
!unzip Data.zip
```

```python
arquivos = sorted(glob('Data/Estados/*.xlsx'))
```

```python
arquivos
```

#Lendo Arquivos para Dataframe

```python
todos_estados = pd.concat((pd.read_excel(cont) for cont in arquivos))
```

```python
todos_estados
```

```python
todos_estados = pd.concat((pd.read_excel(cont) for cont in arquivos), ignore_index=True)
todos_estados
```

```python
todos_estados.to_excel('Aula 2.xlsx', index=False)
```

#Criando Dataframe com Comando Unico

```python
tods_estados = pd.concat((pd.read_excel(cont) for cont in sorted(glob('Data/Estados/*.xlsx'))),ignore_index=True)
todos_estados
```
