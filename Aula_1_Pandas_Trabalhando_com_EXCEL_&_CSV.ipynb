#Lendo Arquivo Excel

```python
uf = pd.read_excel('Estados.xlsx','Estados')
uf
```

#Renomeando Colunas

```python
uf.rename(columns={'UF' : 'Unid Fed'}, inplace = True)
uf
```

#Alterando ordens das colunas

```python
uf.sort_values(by=['Estado'], inplace = True)
uf
```

#Alterando ordens das colunas - Ascendente / Descendente

```python
uf.sort_values(by=['Estado'], inplace = True, ascending = False)
uf
```

#Excluindo colunas

```python
del uf['Sintaxe']
uf
```

#Exportando para excel

```python
uf.to_excel('Aula1_teste.xlsx', sheet_name='AB')
```

#Excluindo coluna Index

```python
uf.to_excel('Aula1_teste.xlsx', sheet_name='AB', index= False)
```
